 ### Q.4: Draw boxplot for Calories vs Category to spot outliers and max calories category

```python
data2=pd.read_csv('menu.csv')
data2.head()
```

```python
data2.Category.value_counts()
```

```python
sns.set(style="whitegrid")
sns.boxplot(
    x="Calories",
    y="Category",
    hue="Category",
    data=data2,
    palette="Set3",
    width=0.2,
    legend=True,
)
plt.xlabel("Calories")
plt.ylabel("Category")
plt.title("Boxplot of Calories by Category")

plt.xticks(rotation=45, ha="right")
plt.show()
```

### Q.5: Figure out which exact item contains a high quantity for ['Calories', 'Total Fat', 'Carbohydrates',  'Dietary Fiber', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calciu 
 (% Daily Value)', 'Iron (% Daily Value). 

```python
for col in columns:
    top_item_valu, top_item_name = data2[[col, 'Item']].nlargest(1, col).values[0]
    print(f"'{col}' : {top_item_name} : `{top_item_valu}`\n")  
```

```python
# ['Calories', 'Total Fat',
#  'Carbohydrates', 'Dietary
#  Fiber', 'Sugars', 'Protein',
#  'Vitamin A (% Daily Value)',
#  'Vitamin C (% Daily Value)',
#  'Calcium (% Daily Value)', 'Iron
#  (% Daily Value)'.
sns.set(style='whitegrid')
plt.figure(figsize=(30,25))

plt.subplot(5, 2, 1)
sns.stripplot(data=data2,x='Category',y='Calories',palette='bright')
plt.title("Calories Content")
plt.xticks(rotation=45,ha='right')

#plot 2:
plt.subplot(5, 2, 2)
sns.stripplot(data=data2,x='Category',y='Total Fat',palette='bright')
plt.title("Total Fat Content")
plt.xticks(rotation=45,ha='right')

plt.subplot(5, 2, 3)
sns.stripplot(data=data2,x='Category',y='Carbohydrates',palette='bright')
plt.title("Carbohydrates Content")
plt.xticks(rotation=45,ha='right')

plt.subplot(5, 2, 4)
sns.stripplot(data=data2,x='Category',y='Dietary Fiber',palette='bright')
plt.title("Dietary Fiber Content")
plt.xticks(rotation=45,ha='right')

plt.subplot(5, 2, 5)
sns.stripplot(data=data2,x='Category',y='Calories',palette='bright')
plt.title("Calories Content")
plt.xticks(rotation=45,ha='right')

plt.subplot(5, 2, 6)
sns.stripplot(data=data2,x='Category',y='Protein',palette='bright')
plt.title("Protein Content")
plt.xticks(rotation=45,ha='right')

plt.subplot(5, 2, 7)
sns.stripplot(data=data2,x='Category',y='Vitamin A (% Daily Value)',palette='bright')
plt.title("Vitamin A (% Daily Value) Content")
plt.xticks(rotation=45,ha='right')

plt.subplot(5, 2, 8)
sns.stripplot(data=data2,x='Category',y='Vitamin C (% Daily Value)',palette='bright')
plt.title("Vitamin C (% Daily Value) Content")
plt.xticks(rotation=45,ha='right')

plt.subplot(5, 2, 9)
sns.stripplot(data=data2,x='Category',y='Calcium (% Daily Value)',palette='bright')
plt.title("Calcium (% Daily Value) Content")
plt.xticks(rotation=45,ha='right')

plt.subplot(5, 2, 10)
sns.stripplot(data=data2,x='Category',y='Iron (% Daily Value)',palette='bright')
plt.title("Iron (% Daily Value) Content")
plt.xticks(rotation=45,ha='right')

plt.show()
```
